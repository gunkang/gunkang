```
import pandas as pd
import numpy as np 
```

```
##Converting Tab delimited file to csv   #import
import csv
df = []
with open('Thomas_TX_imports.txt', 'r',encoding = "ISO-8859-1") as f:
    reader = csv.reader(f, dialect='excel', delimiter='\t')
    for row in reader:
      df.append(row)

    df = pd.DataFrame(df)
    print(df)
```

```
df.to_csv('i1.csv')
```

```
##Converting Tab delimited file to csv   #export
import csv
df1 = []
with open('Thomas_TX_exports.txt', 'r',encoding = "ISO-8859-1") as f:
    reader = csv.reader(f, dialect='excel', delimiter='\t')
    for row in reader:
      df1.append(row)

    df1 = pd.DataFrame(df1)
    print(df1)
```

```
#export file to csv
df1.to_csv('e1.csv')
```

```
#importing thomas_imports/exports csv data
dfi = pd.read_csv('i1.csv', low_memory=False, header= 1)
dfe = pd.read_csv('e1.csv', low_memory=False, header= 1)
```

```
dfi[['DateTime1','TimeZone']] = dfi.eventtimestamp.str.split(" G",expand=True) 

#splittung eventstamp column
```

```
dfi[['Day','Month','Date','year','time' ]] = dfi.DateTime1.str.split(" ",expand=True)
```

```
dfi.Month.value_counts()
```

```
#replacing strings with int for creating date column
dfi = dfi.replace(regex={r'^Jan$': '01', 'Feb': '02','Mar':'03','Apr':'04','May':'05','Jun':'06',"Jul":'07','Aug':'08','Sep':'09','Oct':'10','Nov':'11',"Dec":'12'})
```

```
dfi['Date'] = dfi["Month"]+"/"+dfi["Date"]+"/"+dfi["year"]
```

```
dfi['DateTime'] = dfi['Date'] +" "+ dfi["time"]
```

```
dfi = dfi.drop(['0','eventtimestamp','Day','Month','year','DateTime1','time'],axis=1)
```

```
dfi.sellingtaxonomy.value_counts()
```

```
dfi.to_csv('import_data.csv',header=True)
```
